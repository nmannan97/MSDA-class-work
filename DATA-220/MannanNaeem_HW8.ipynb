```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

```python
# function and its derivative
def f(x):
    return 2*x**2 - x + 2

def df(x):
    return 4*x - 1   # derivative of 2x^2 - x + 2
```

```python
# hyperparameters
x0 = 3.0          # starting point (you can change this)
lr = 0.1          # learning rate
epochs = 30       # number of iterations

x = x0
history_x = [x0]
history_f = [f(x0)]

for i in range(epochs):
    grad = df(x)
    x = x - lr * grad
    history_x.append(x)
    history_f.append(f(x))

print("Final x from gradient descent:", x)
print("Final f(x):", f(x))
```

```python
# make a smooth curve for f(x)
xs = np.linspace(-1, 3.5, 200)
ys = f(xs)

plt.figure(figsize=(7, 5))
plt.plot(xs, ys, label="f(x) = 2x^2 - x + 2")

# plot the GD steps
plt.scatter(history_x, history_f, color="red", zorder=5, label="GD steps")
plt.plot(history_x, history_f, color="red", linestyle="--", alpha=0.6)

for i, (xx, yy) in enumerate(zip(history_x, history_f)):
    plt.text(xx, yy+0.1, str(i), fontsize=8)

plt.title("Gradient Descent on f(x) = 2xÂ² - x + 2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python

# objective for scipy (must take array-like sometimes, but here 1D is fine)
def f_scipy(x):
    return 2*x**2 - x + 2

def df_scipy(x):
    return 4*x - 1

# initial guess
x0 = np.array([3.0])

res = minimize(f_scipy, x0, jac=df_scipy, method='BFGS')

print("SciPy result x* =", res.x[0])
print("SciPy f(x*)    =", res.fun)
```
