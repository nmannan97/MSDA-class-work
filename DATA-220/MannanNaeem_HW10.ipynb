## Part 1: Python Function Implementation

The following function computes the posterior probability of having a disease given a positive diagnostic test result. It uses **Bayes’ Theorem**, which combines the prior probability of having the disease with the characteristics of the test (sensitivity and specificity).

```python

````python
def posterior_probability(prior, sensitivity, specificity):
    """
    Calculate P(Disease | Positive Test) using Bayes' Theorem.

    prior: P(Disease)
    sensitivity: P(Positive | Disease)
    specificity: P(Negative | No Disease)
    """

    # False positive rate
    false_positive_rate = 1 - specificity

    # Bayes numerator: P(Positive | Disease) * P(Disease)
    numerator = sensitivity * prior

    # Bayes denominator:
    # P(Positive | Disease)*P(Disease) + P(Positive | No Disease)*P(No Disease)
    denominator = numerator + false_positive_rate * (1 - prior)

    return numerator / denominator
````

## Part 2: Explanation of the Code

We want to calculate the probability that a person actually has the disease *given* that they tested positive. This is the **posterior probability**, and it is computed using **Bayes’ Theorem**:

\[
P(D|+) = \frac{P(+|D)P(D)}{P(+|D)P(D) + P(+|\neg D)P(\neg D)}
\]

Where:

- \( P(D) \) is the prior probability of the disease (prevalence).
- \( P(+|D) \) is the sensitivity or true positive rate.
- \( P(\neg D) = 1 - P(D) \) is the probability of not having the disease.
- \( P(+|\neg D) = 1 - \text{specificity} \) is the false positive rate.

The function `posterior_probability()` directly implements this formula:

1. Computes the false positive rate.  
2. Computes the numerator of Bayes’ formula.  
3. Computes the denominator (total probability of testing positive).  
4. Returns the posterior probability.  

````python
# Given values
prior = 1/10000            # Disease prevalence: 1 in 10,000
sensitivity = 0.99         # True positive rate
specificity = 0.95         # True negative rate

result = posterior_probability(prior, sensitivity, specificity)
print("Posterior probability of disease given a positive test:", result)
````
