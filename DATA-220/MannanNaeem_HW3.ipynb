```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Files/grade.csv")
df.head()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x="grade", hue="msex", bins=5, multiple="dodge", edgecolor="black")

plt.title("Distribution of Grades by Gender")
plt.xlabel("Grade")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.boxplot(data=df, x="msex", y="grade")

plt.title("Grades by Gender")
plt.xlabel("Gender")
plt.ylabel("Grade")
plt.show()
```
