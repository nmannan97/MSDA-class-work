```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import numpy as np
```

```python
df1 = pd.read_csv("USW00014819.csv")
df2 = pd.read_csv("USC00042319.csv")

df3 = pd.read_csv("USW00012918.csv")
df4 = pd.read_csv("USW00093107.csv")
```

```python
plt.figure(figsize=(10, 5))

plt.plot(df3['DATE'], df1["DLY-TAVG-NORMAL"], label='Station 1')
plt.plot(df3['DATE'], df2["DLY-TAVG-NORMAL"], label='Station 2')
plt.plot(df3['DATE'], df3["DLY-TAVG-NORMAL"], label='Station 3')
plt.plot(df3['DATE'], df4["DLY-TAVG-NORMAL"], label='Station 4')

plt.xlabel('Date')
plt.ylabel('Daily Average Temperature (°F)')
plt.title('Daily Average Temperature Normal by Station')

step = 12  # show every 10th date

plt.xticks(
    df3['DATE'][::step],
    df3['DATE'][::step],
    rotation=45
)

plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
